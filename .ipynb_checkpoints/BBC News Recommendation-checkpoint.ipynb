## BBC News Recommendation System

```python
# BBC News Recommendation System Demo

# Import required libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to load and preprocess the data
def load_data(file_path):
    """
    Load and preprocess the BBC News dataset.

    Args:
        file_path (str): Path to the CSV file containing the BBC News dataset.

    Returns:
        pd.DataFrame: A DataFrame containing a random sample of 500 articles.
    """
    df = pd.read_csv(file_path, sep='\t').sample(500, random_state=115)
    return df

# Function to recommend articles
def recommend_articles(user_query, df, top_n=5):
    """
    Recommend articles based on user input.

    Args:
        user_query (str): The user's search query or topic.
        df (pd.DataFrame): The preprocessed dataset.
        top_n (int): Number of top recommendations to return.

    Returns:
        list: Top N recommended articles with title, category, and similarity score.
    """
    # Combine titles and content for TF-IDF vectorization
    df['combined'] = df['title'] + " " + df['content']
    
    # Initialize the TF-IDF vectorizer
    vectorizer = TfidfVectorizer(stop_words='english')
    
    # Fit and transform the combined text
    tfidf_matrix = vectorizer.fit_transform(df['combined'])
    
    # Transform the user query
    user_vec = vectorizer.transform([user_query])
    
    # Calculate cosine similarities
    similarities = cosine_similarity(user_vec, tfidf_matrix).flatten()
    
    # Get the indices of top N recommendations
    top_indices = similarities.argsort()[-top_n:][::-1]
    
    # Prepare the results
    results = [
        (df.iloc[idx]['title'], df.iloc[idx]['category'], similarities[idx])
        for idx in top_indices
    ]
    return results

# Load dataset
file_path = 'bbc-news-data.csv'
data = load_data(file_path)

# Display dataset structure
print("Dataset preview:")
data.head()
```

```python
# User interaction
print("\nWelcome to the BBC News Recommendation System!")
user_query = input("Enter a news topic or description: ")

# Generate recommendations
recommendations = recommend_articles(user_query, data)

# Display the recommendations
print("\nTop 5 Recommended News:")
for title, category, similarity in recommendations:
    print(f"-> {title} (Category: {category}, Similarity: {similarity:.2f})")
```

```python
# Explain TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_sample = vectorizer.fit_transform(data['combined'][:5])

# Show feature names
print("TF-IDF Features Example:")
print(vectorizer.get_feature_names_out()[:10])

# Show TF-IDF matrix for the first 5 articles
pd.DataFrame(tfidf_sample.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize similarities for the top 5 recommended articles
similarities = [sim for _, _, sim in recommendations]
titles = [title for title, _, _ in recommendations]

plt.figure(figsize=(10, 6))
sns.barplot(x=similarities, y=titles, palette="viridis")
plt.xlabel("Similarity Score")
plt.ylabel("Article Title")
plt.title("Top 5 Recommended Articles")
plt.show()
```
